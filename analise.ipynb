### **Questão 02** - No modelo com memoria (AR(p)) com defasagem, levando em consideração o critério de Estacionariedade. Calcule a variância (y(t)) e a covariância (y(t) com y(t-1)) da série quando alfa(0)=0 e u=0 interprete.

**Resposta:**

Para um modelo AR(1): $y_t = \alpha_1y_{t-1} + \epsilon_t$ com $\alpha_0 = 0$ e $\mu = 0$

**Cálculo da Variância $\gamma(0)$:**

1) Partindo da equação: $y_t = \alpha_1y_{t-1} + \epsilon_t$

2) A variância $\gamma(0)$ é dada por:
    $\gamma(0) = E(y_t^2)$

3) Multiplicando a equação por $y_t$:
    $y_t^2 = \alpha_1y_ty_{t-1} + y_t\epsilon_t$

4) Aplicando o operador esperança:
    $E(y_t^2) = \alpha_1E(y_ty_{t-1}) + E(y_t\epsilon_t)$
    $\gamma(0) = \alpha_1\gamma(1) + \sigma_\epsilon^2$

5) Como $\gamma(1) = \alpha_1\gamma(0)$, substituindo:
    $\gamma(0) = \alpha_1^2\gamma(0) + \sigma_\epsilon^2$

6) Resolvendo para $\gamma(0)$:
    $\gamma(0)(1-\alpha_1^2) = \sigma_\epsilon^2$
    $\gamma(0) = \frac{\sigma_\epsilon^2}{1-\alpha_1^2}$

**Cálculo da Covariância $\gamma(1)$:**

1) A covariância $\gamma(1)$ é dada por:
    $\gamma(1) = E(y_ty_{t-1})$

2) Multiplicando a equação por $y_{t-1}$:
    $y_ty_{t-1} = \alpha_1y_{t-1}^2 + \epsilon_ty_{t-1}$

3) Aplicando o operador esperança:
    $E(y_ty_{t-1}) = \alpha_1E(y_{t-1}^2) + E(\epsilon_ty_{t-1})$
    $\gamma(1) = \alpha_1\gamma(0)$

4) Substituindo $\gamma(0)$:
    $\gamma(1) = \alpha_1\frac{\sigma_\epsilon^2}{1-\alpha_1^2}$

**Interpretação:**
- A variância $\gamma(0)$ depende de $\sigma_\epsilon^2$ e $\alpha_1$
- Para que o processo seja estacionário, é necessário que $|\alpha_1| < 1$
- A covariância $\gamma(1)$ é proporcional à variância, com fator $\alpha_1$
